## Linear sample data 

### Creation of sample linear data (with some random noise)

#### Import the required libraries.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

#### Random 10k numbers of X values between 0 and 10 .

```python
X = np.linspace(0,10,10000)
```

#### Initialize Slope and Intercept.

```python
m = 0.5
b = 5
```

#### Linear Equation

```python
y = m*X + b
```

#### Let's add some noise to it

```python
noise = np.random.randn(len(X))
base_y = y
y = y + noise
```

#### Convert the numpy arrays into pandas dataframe , so that we can do some operation such as merging etc.

```python
Xdf = pd.DataFrame(X, columns = ['X'])
ydf = pd.DataFrame(y, columns = ['y'])
zdf = pd.concat([Xdf,ydf], axis = 1)
```

#### Plot the data 

```python
sample_size = 500
sample = list(np.random.randint(0,len(zdf), sample_size))
zdf_sample = zdf.iloc[sample]
zdf_sample.plot(kind='scatter',x='X',y='y')
plt.plot(X,base_y,'r')
```

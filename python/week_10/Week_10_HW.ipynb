````python
# Praba Santhanakrishnan
import json
import csv
import pickle

# Load the data from json file
with open("data.json","rt") as json_data_file:
    json_data = json.loads(json_data_file.read())

# Load the data from csv file
with open("data.csv","rt") as csv_data_file:
    csv_reader = csv.reader(csv_data_file)
    csv_data = []
    for line in csv_reader:
        csv_data.append(line)

# Load the data from pickle file
with open("data.pkl","rb") as pkl_data_file:
    pkl_data = pickle.load(pkl_data_file)
````

````python
# Dictionary to hold the data in the standard format from all three data sources.

people = {} 

# Store the data from csv in the standard dictionay format, for example:

# {'key':{'Name':'XXX','Phone': 'XXX-XXX-XXXX','Address': '123 Abc Road','City': 'Seattle','Country':'United States','Email:'abc@xyz.com'}}


for item in csv_data:
    people[item[0]] = {csv_data[0][1]:item[1],csv_data[0][2]:item[2],csv_data[0][3]:item[3],csv_data[0][4]:item[4],csv_data[0][5]:item[5],csv_data[0][6]:item[6]}

# Remove the first item which is a label
people.pop('')

# Store the data from json in the standard dictionay format, for example:
# {'key':{'Name':'XXX','Phone': 'XXX-XXX-XXXX','Address': '123 Abc Road','City': 'Seattle','Country':'United States','Email:'abc@xyz.com'}}


for key1,value1 in json_data.items():
    for key2,value2 in value1.items():
        if str(key2) in people.keys():
            dict_temp = people[str(key2)]
                    
        else: dict_temp = {}
        dict_temp[key1] = value2
        people[str(key2)] = dict_temp

    
# Store the data from pkl in the standard dictionay format, for example:
# {'key':{'Name':'XXX','Phone': 'XXX-XXX-XXXX','Address': '123 Abc Road','City': 'Seattle','Country':'United States','Email:'abc@xyz.com'}}

for item in pkl_data.items():
    for key , value in item[1].items():
        
        if str(key) in people.keys():
            dict_temp = people[str(key)]
            
        else: dict_temp = {}
        dict_temp[item[0]] = value
        people[str(key)] = dict_temp
        
#***********************************************************************************        
# 1.What are the unique countries in the dataset, sorted alphabetically?
#***********************************************************************************
# convert to set to automatically include only the unique items
set_unique_country = sorted(set([item[1]['Country'] for item in people.items()]))

# write to the file for question 1
with open("question_1.csv","wt") as q1_file:
    for country in set_unique_country:
        q1_file.write(country + '\n')
        
    

#***********************************************************************************        
#2.What are the unique email domains in the dataset, sorted alphabetically?
#***********************************************************************************

import re

# set_unique_email = sorted(set([item[1]['Email'] for item in people.items()]))
list_email_domain = []

for item in people.items():
    domain = re.search("(?<=@)[^.]+.+",item[1]['Email'])   # Reg ex for extracting domain name only
    list_email_domain.append(domain.group())

# convert to set to automatically include only the unique items
set_unique_email = sorted(set(list_email_domain), key=str.lower)

# write to the file for question 2
with open("question_2.csv","wt") as q2_file:
    for email in set_unique_email:
        q2_file.write(email + '\n')

        
#***********************************************************************************        
#3.First Names where address is not P.O Box 
#***********************************************************************************
list_first_name = []
for item in people.items():    

    if bool(re.match("P.O. Box", item[1]['Address'])) is False: # Reg ex for checking if the address contain P. O Box
        domain = re.search('^([A-Z][a-z-]+)',item[1]['Name']) # Reg ex for extracting first name only
        list_first_name.append(domain.group())
        
list_first_name = sorted(list_first_name)


# write to the file for question 3
with open("question_3.csv","wt") as q3_file:
    for name in list_first_name:
        q3_file.write(name + '\n')

        
#***********************************************************************************  
# 4. What are the names of the first 5 people when you sort the data alphabetically by Country?
#***********************************************************************************  
import operator

#dict_name_country = {}
list_name_country = []
for item in people.items():         
    list_name_country.append([item[1]['Name'], item[1]['Country']])

# sort by the country
sorted_list_country = sorted(list_name_country, key=operator.itemgetter(1))

#pick top 5 names
sorted_list_country = sorted_list_country[0:5]
    
# write to the file for question 4
with open("question_4.csv","wt") as q4_file:
    for name in sorted_list_country:
        q4_file.write(name[0] + '\n') 
        

#***********************************************************************************  
# 5. What are the names of the first 5 people when you sort the data numerically ascending by phone number?
#***********************************************************************************  
import operator


list_name_phone = []
for item in people.items():         

    list_name_phone.append([item[1]['Name'],item[1]['Phone']])
    

# sort by the phone
sorted_list_phone = sorted(list_name_phone, key=operator.itemgetter(1))

#pick top 5 names
sorted_list_phone = sorted_list_phone[0:5]
    
# write to the file for question 4
with open("question_5.csv","wt") as q5_file:
    for name in sorted_list_phone:
        q5_file.write(name[0] + '\n') 

````

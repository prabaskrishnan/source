## 1. List, Mapping, Filtering, and Reducing Comprehensions

**A.** Write a function that doubles an integer or floating type values, reverses the content if it is a string, and does nothing if it is another kind of value.

Use that function in a **list comprehension** and apply it to the below list of data.

Example of the first two entries:
```
['cinesra', 10, ...]
```

````python
fix_me = ['arsenic', 5, 'ball', 2, 47, 123, 
          'comprehensions', 'list', 5.5, 
          8343.23, 'shishkabob', True]
````

````python
# Put the function here
def func_map(value):
    if type(value) is int or type(value) is float:
        return value * 2
    elif type(value) is str:
        return value[::-1]
    else: return value
````

````python
# Put the list comphrension (which calls your function above) here 
fixed = [func_map(item) for item in fix_me]

print(fixed)
````

**B.** Now apply your function using a **map**.

````python
fixed = map(func_map,fix_me)
print(list(fixed))
````

**C.** Repeat step **B**, except use **filter** to remove any number values that are less than 5 after your map.  You may use a separate function to filter (or could do it with one statement).

````python
def func_filter(value):
    if type(value) is int or type(value) is float:
        if value < 5: return False
    return True

fixed = filter(func_filter , map(func_map,fix_me))
print(list(fixed))
````


````python
fixed = filter(lambda x: False if type(x) is int and x < 5 else  False if type(x) is float and x < 5.0 else True , map(func_map,fix_me))
print(list(fixed))
````
